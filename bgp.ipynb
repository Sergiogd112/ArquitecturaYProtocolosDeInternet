```python
class BGP_table:
    def __init__(self):
        self.table = pd.DataFrame(
            columns=[
                "selected",
                "Network",
                "Mask",
                "Next Hop",
                "Metric",
                "LocPrf",
                "Weight",
                "Path",
                "Removed"
            ]
        )

    def update(self, pack):
        print("+++")
        bgp_pack = pack[BGPUpdate]
        while bgp_pack:
            print(bgp_pack)
            # Console().print(Panel(Pretty(bgp_pack.fields)))
            path_attr = bgp_pack[BGPPathAttr]
            Console().print(Panel(Pretty(path_attr.fields)))
            # check if it has a payload
            if bgp_pack.payload:
                bgp_pack = bgp_pack.payload[BGPUpdate]
            else:
                break
    def to_rich_table(self):
        table=Table()
        table.add_column("Selected")
        table.add_column("Network")
        table.add_column("Mask")
        table.add_column("Next Hop")
        table.add_column("Metric")
        table.add_column("LocPrf")
        table.add_column("Weight")
        table.add_column("Path")
        for index, row in self.table.iterrows():
            table.add_row(
                str(row["selected"]),
                row["Network"],
                row["Mask"],
                row["Next Hop"],
                row["Metric"],
                row["LocPrf"],
                row["Weight"],
                row["Path"],
            )
        return table
```
